Exemple d'une différence de Gaussiennes

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

Lecture d'une image et affichage.

```python
image = cv2.imread('RGB.jpg',cv2.IMREAD_GRAYSCALE)
plt.figure(figsize = (10,10))
plt.imshow(image,cmap = plt.get_cmap('gray'))
plt.show()
```

Convolution de l'image avec deux filtres Gaussiens d'écart-types différents. Le deuxième filtre rend l'image plus flou.

```python
convavecGauss1 = cv2.GaussianBlur(image.astype(float),(5,5),3)
convavecGauss2 = cv2.GaussianBlur(image.astype(float),(5,5),5)

f, axarr = plt.subplots(1, 2, figsize=(18,15))
axarr[0].imshow(convavecGauss1.astype('uint8'),cmap = plt.get_cmap('gray'))
axarr[1].imshow(convavecGauss2.astype('uint8'),cmap = plt.get_cmap('gray'))
plt.show()
```

La différence de Gaussiennes est la différence des deux images convoluées avec les filtres Gaussiens. 

```python
difGauss = np.absolute(convavecGauss1.astype(float)-convavecGauss2.astype(float)) #Fait le calcul en float.

difGauss = difGauss * 255/np.max(difGauss) #Mise à l'échelle pour affichage
plt.figure(figsize = (10,10))
plt.imshow(difGauss.astype('uint8'),cmap = plt.get_cmap('binary')) #Plus foncé = valeur plus élevée
plt.show()
```
