Exemple du calcul de filtres de Gabor

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

Lecture de l'image en tons de gris et affichage. 

```python
image = cv2.imread('bureau.jpg',cv2.IMREAD_GRAYSCALE)
plt.figure(figsize = (7,7))
plt.imshow(image,cmap = plt.get_cmap('gray'))
plt.show()
```

Création du filtre de Gabor avec cv2.getGaborKernel(ksize, sigma, theta, lambda, gamma, psi, ktype)

ksize: size of gabor filter (n, n), sigma: standard deviation of the gaussian, theta: orientation of filter, lambda: wavelength of the sinus, gamma: spatial aspect ratio, psi: phase offset

```python
Filtre  = cv2.getGaborKernel((9,9), 2.0, 0.0, 4.0, 1.0, np.pi/2, ktype=cv2.CV_32F)
plt.imshow(Filtre, cmap = plt.get_cmap('gray'))
plt.show()
print(Filtre)
```

Application du filtre sur l'image

```python
fimg = cv2.filter2D(image, cv2.CV_8UC3, Filtre)
plt.figure(figsize = (7,7))
plt.imshow(fimg, cmap = plt.get_cmap('gray'))
plt.show()
```
