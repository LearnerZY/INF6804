Exemple du calcul d'une approximation polygonale

```python
import cv2
from skimage import data
import numpy as np
import matplotlib.pyplot as plt
```

Lecture et affichage d'une image provenant de skimage.

```python
image = np.invert(data.horse())
image = image.astype('uint8')
plt.figure(figsize=(6,6)) 
plt.imshow(image, plt.get_cmap('binary'))
plt.show()
```

Extraction du contour externe. Pour l'affichage, on extrait les informations de la structure de contour, et on les met dans deux listes pour X et Y.

```python
im2, contours, hierarchy = cv2.findContours(image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

contx = []
conty = []
for elem in enumerate(contours[0]):
    contx.append(elem[1][0][0]) #il y a plusieurs niveaux de liste...
    conty.append(elem[1][0][1])   
```

Affichage du contour sur l'image originale.

```python
plt.figure(figsize=(6,6)) 
plt.imshow(image, plt.get_cmap('binary'))
plt.plot(contx,conty,'r')
plt.show()
```

Approximation polygonale avec une erreur choisie. Affichage.

```python
erreur = 10
approxP=cv2.approxPolyDP(contours[0], erreur, True)

approxx = []
approxy = []
for elem in enumerate(approxP):
    approxx.append(elem[1][0][0])
    approxy.append(elem[1][0][1]) 
    
plt.figure(figsize=(6,6)) 
plt.imshow(image, plt.get_cmap('binary'))
plt.plot(approxx,approxy,'r')
plt.show()    
```
