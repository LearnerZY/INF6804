Exemple du calcul de pyramide Gaussiennes et Laplaciennes

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

Lecture de l'image en tons de gris et affichage. 

```python
image = cv2.imread('bureau.jpg',cv2.IMREAD_GRAYSCALE)
plt.figure(figsize = (7,7))
plt.imshow(image,cmap = plt.get_cmap('gray'))
plt.show()
```

Calcul de la pyramide Gaussienne avec la fonction pyrDown(). Application d'un filtre Gaussien suivi d'un sous-échantillonage. 

```python
G = image.copy()
gpImage = [G]
for i in range(6):
    G = cv2.pyrDown(G, borderType = cv2.BORDER_REPLICATE)
    gpImage.append(G)
    plt.figure(figsize = (5,5))
    plt.imshow(G,cmap = plt.get_cmap('gray'))
    plt.show()
```

Calcul de la pyramide Laplacienne. C'est la différence entre l'image de la pyramide Gaussienne et une image agrandie avec pyrUp(). Les différences proviennent du flou Gaussien.

```python
L = image.copy()
lpImage = [gpImage[5]]
plt.imshow(gpImage[5],cmap = plt.get_cmap('gray'))
plt.show()
for i in range(5,0,-1):
    GE = cv2.pyrUp(gpImage[i])
    L = cv2.subtract(gpImage[i-1],GE)
    lpImage.append(L)
    plt.figure(figsize = (5,5))
    plt.imshow(L,cmap = plt.get_cmap('gray'))
    plt.show()
```
