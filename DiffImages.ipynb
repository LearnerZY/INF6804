Différence de deux images dans une séquence vidéo

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

Lecture de deux trames

```python
image1 = (cv2.imread('in000235.jpg',cv2.IMREAD_GRAYSCALE)).astype(float)
plt.imshow(image1, cmap = plt.get_cmap('gray'))
plt.show()
image2 = (cv2.imread('in000240.jpg',cv2.IMREAD_GRAYSCALE)).astype(float)
plt.imshow(image2, cmap = plt.get_cmap('gray'))
plt.show()
```

Différence des deux trames, et affichage des différences

```python
Diffimage = np.abs(image1 - image2)
fig1 = plt.figure(figsize = (10,10))
plt.imshow(Diffimage, cmap = plt.get_cmap('gray'))
plt.show()
```

Notez qu'on détecte des changements à la nouvelle et à l'ancienne position.
